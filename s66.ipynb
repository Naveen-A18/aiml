```python
#to find whether a given number is prime or not
n = int(input("Enter a number: "))
for i in range (2,n):
    if(n%i==0):
        print("not prime")
        break
else:
        print("prime")
```

```python
#to calculate the factorial of a given numbers using for loop
n = int(input("enter the number:"))
pr = 1
for i in range (1,n+1):
    pr=pr*i
print(i,pr)
print(f"fact is {n},iteration{i},{pr}")
```

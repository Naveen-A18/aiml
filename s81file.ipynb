```python
#file creating
file = open("demo1.text","w")
file.write("this is file handling in python\n")
file.write("this is AI\n")
print("file is created")
file.close()
```

```python
#list of nodes
f = open("demo2.text","w")
list=["apple\n","mango\n","grapes\n"]
f.writelines(list)
f.close()
```

```python
#read
f =open("demo1.text","r")
file_conten =f.read(15)
print(file_conten)
```

```python
#read line #to read multiple
file=open("demo1.text","r")
con = file.readlines()
print(con)
```

```python
#using the  with keyword
with open("demo1.text","r")as file:
    content = file.read()
    print(content)
#you no need to clode the file while using the with the statement
```

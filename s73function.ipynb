```python
def fact(num):
    reslut = 1 
    for i in range (1,num+1):
        reslut = reslut*i
    return reslut
print(fact(7))
```

```python
#fibonacci series
def fibonacci(n):
    fib_sequence = [0,1]
    while len(fib_sequence)<n:
        fib_sequence.append(fib_sequence[-1]+fib_sequence[-2])
    return fib_sequence
print(fibonacci(5))
```

```python
#recursion
#fuction will call to itself
def factorial(n):
    if n==0 or n==1 :
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
print(factorial(1))
print(factorial(0))
```

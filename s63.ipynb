```python
#shyam is reading a mobile number as words and he went to copy that in the form of numeric values
#ex-nine four two three zero seven.
#result -942307
words_to_digits = {
    "zero": "0",
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9"
}
mobile_words = input("Enter mobile number in words: ")  
parts = mobile_words.split()
result = ""
for word in parts:
    result += words_to_digits[word.lower()]
print("Numeric mobile number:", result)
```

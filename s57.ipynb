```python
# to take single digit number from the key board and print is value in english word?
num = int(input("Enter a single digit number (0-9): "))

words = {
    0: "Zero",
    1: "One",
    2: "Two",
    3: "Three",
    4: "Four",
    5: "Five",
    6: "Six",
    7: "Seven",
    8: "Eight",
    9: "Nine"
}

if num in words:
    print("In English:", words[num])
else:
    print("Please enter a valid single digit number!")
```

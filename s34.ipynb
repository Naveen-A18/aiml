```python
#to store seven fruits in a list entered by the user
fruits= []
for i in range(7):
    fruit= input("Enter the fruit name: ")
    fruits.append(fruit)    
print("Fruits you entered are: ", fruits)
```

```python
#to accpect marks of 6 students and display them in a sorted manner
marks= []
for i in range(6):
    mark= int(input("Enter the marks of student {}: "))
    print("enter the student marks",mark)
    marks.append(mark)
marks.sort()
print("Sorted marks are: ", marks)
```
